```
import pandas as pd
from matplotlib import pyplot as plt
pd.set_option('display.max_columns', None)  # or 1000
pd.set_option('display.max_rows', None)  # or 1000
pd.set_option('display.max_colwidth', -1)  # or 199
```

```
kader = pd.read_csv('kader.csv')
pd.set_option('display.max_columns', None)  
```

```
proditp = kader[kader['Prodi'] == 'Teknologi Pangan']
proditif = kader[kader['Prodi'] == 'Teknik Informatika']
prodite = kader[kader['Prodi'] == 'Teknik Elektro']
proditk = kader[kader['Prodi'] == 'Teknik Kimia']
proditi = kader[kader['Prodi'] == 'Teknik Industri']
prodi = [proditp.count()[0], proditif.count()[0], prodite.count()[0], proditk.count()[0], proditi.count()[0]]
```

```
prodi_name = ['Teknologi Pangan', 'Teknik Informatika', 'Teknik Elektro', 'Teknik Kimia', 'Teknik Industri']
fig, axes = plt.subplots()

print('Jumlah Kader \n\n\n')

for i in range(len(prodi_name)):
  print(prodi_name[i], " : ", prodi[i]);

print('Jumlah Kader : ', kader.count()[0])

for i in range(len(prodi_name)):
  plt.bar(prodi_name[i], prodi[i]);

plt.xlabel('Prodi')
plt.ylabel('Jumlah')
fig.set_size_inches(12, 5)
```
